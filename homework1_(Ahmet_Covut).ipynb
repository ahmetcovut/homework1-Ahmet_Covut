```python
value= ahmcvt.fmarno
histogram=thinkstats2.Hist()
histogram2= thinkstats2.Hist(value,label="marriage" )
thinkplot.Hist(histogram2)
thinkplot.Config(xlabel="age",ylabel="pregnant")
```

```python
a=ahmcvt[ahmcvt.fmarno == 0]
b = ahmcvt[ahmcvt.fmarno >= 1]
a_histogram = thinkstats2.Hist(a.totincr, label="NeverMarried")
b_histogram = thinkstats2.Hist(b.totincr, label="Married")
thinkplot.Hist(a_histogram, align="left",width=0.5)
thinkplot.Hist(b_histogram,align="right",width=0.5)
thinkplot.Config(xlabel="WEEKS", ylabel="COUNTS")
```
